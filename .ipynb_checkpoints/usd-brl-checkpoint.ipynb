# Análise de USD (dólar) x BRL (reais)

A hipótese inicial é que exista uma diferença (para menor) no valor de conversão de USD para BRL entre o 5º e o 10º dia útil de cada mês.
O motivo é que nesse período deve entrar uma quantidade maior de USD no Brasil dado que exista a prestação de serviços para o exterior (ou até mesmo outros contratos que devem ter pagamento nessa data).

Para validar a ideia, será utilizado o valor de conversão histórico desde 2015-01-01 até 2019-12-19 (data atual).
Os dados foram obtidos pelo _Google Finance_ através de uma planilha no _Google Spreasheets_ com a função `=GOOGLEFINANCE("USDBRL", "price", "01/01/2015", "12/18/2019", "DAILY")`.

```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df_usd_brl = pd.read_csv(
    './data/usd_brl.csv',
    names=['data', 'valor'],
    header=0,
    parse_dates=['data'],
    index_col='data',
)
df_usd_brl.index = df_usd_brl.index.normalize()
df_usd_brl.head()
```

```python
df_feriados = pd.read_excel('http://www.anbima.com.br/feriados/arqs/feriados_nacionais.xls')
df_feriados.tail(10)
```

```python
# remove os valores de 936-944
df_feriados = df_feriados.drop(range(936, 945), 'index')
```

```python
df_feriados.head()
```

```python
# remove feriados
# erros ignorados porque existem datas em `datas_feriados` que não existem em `df_usd_brl`
datas_feriados = pd.to_datetime(df_feriados.Data)
df_usd_brl = df_usd_brl.drop(datas_feriados, errors='ignore')
```

```python
# remove finais de semana
dias_da_semana = set(range(0, 5))
df_usd_brl = df_usd_brl[df_usd_brl.index.weekday.isin(dias_da_semana)].copy()
```

```python
df_usd_brl['ano'] = df_usd_brl.index.year
df_usd_brl['mes'] = df_usd_brl.index.month
df_usd_brl['dia'] = df_usd_brl.index.day
meses = df_usd_brl.mes.astype(str).str.rjust(2, '0')
df_usd_brl['ano_mes'] = df_usd_brl.ano.astype(str).str.cat(meses, sep='_')
df_usd_brl.head()
```

```python
# computa o dia útil sequencial no mês
df_usd_brl['dia_util'] = df_usd_brl.groupby('ano_mes').dia.rank()
```

```python
(fig, ax) = plt.subplots(1, 1, figsize=(15, 10))
sns.boxplot(df_usd_brl.dia_util, df_usd_brl.valor, ax=ax)
ax.set_xlabel('Dia útil')
ax.set_ylabel('Valor dólar')
ax.set_title('Valor do dólar por dia útil entre 2015-2019')
```

Pelo gráfico acima, aparentemente existe uma queda no valor da cotação de USD para BRL entre o 5º e 8º dia útil de cada mês.

A hipótese a ser testada é:

$H_{0}$: A média no 4º dia útil é menor ou igual a média do 5º dia útil ($\mu_{4} \le \mu_{5}$)

$H_{A}$: A média no 4º dia útil é maior que a média do 5º dia útil ($\mu_{4} \gt \mu_{5}$)

```python
a = df_usd_brl[df_usd_brl.dia_util == 4].valor.values
b = df_usd_brl[df_usd_brl.dia_util == 5].valor.values
(t, p) = stats.ttest_ind(a, b)

print('t', t)
print('p', p / 2)
# precisa dividir por 2 para ser o teste de uma calda
# para mais informações https://stackoverflow.com/a/15984310
```

Aparentemente, o resultado de uma média um pouco menor no 5º dia útil é ocasional.

Como esperado `t < 0` (5º dia é menor que o 4º), mas `p / 2` deveria ser menor 0.025 (no mínimo, considerando 5%).

Portanto, aceitamos $H_{0}$

Outra possibilidade de teste é assumir que, sendo um valor aleatório, $P(X_5 < X_4) = 0.5$ (probabilidade do 5º < 4º é 50%).

Assim, é possível modelar o teste com uma comparação com uma distribuição binomial.

$H_{0}$: $P(X_5 \lt X_4) \le 0.5$

$H_{A}$: $P(X_5 \lt X_4) \gt 0.5$

```python
dia_4 = df_usd_brl[df_usd_brl.dia_util == 4].valor.values
dia_5 = df_usd_brl[df_usd_brl.dia_util == 5].valor.values

sucessos = (dia_5 < dia_4).sum()
p = stats.binom_test(sucessos, n=len(dia_5), p=0.5, alternative='greater')
print('p', p)
```

A $H_{A}$ também não é válida nesse caso.

Portanto, aceitamos $H_0$.

Assim, a conclusão é que não existe uma diferença significativa no valor do dólar para o 5º dia útil, mesmo que a média histórica seja um pouco menor.
