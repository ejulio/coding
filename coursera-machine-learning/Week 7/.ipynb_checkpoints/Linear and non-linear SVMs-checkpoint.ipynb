```python
% matplotlib inline
import matplotlib.pyplot as plt
from scipy.io import loadmat
import numpy as np
from sklearn.svm import LinearSVC, SVC
```

```python
data = loadmat('ex6data1.mat')
X = data['X']
y = np.int32(data['y']).squeeze()
colors = ['white' if v == 1 else 'black' for v in y]
plt.scatter(X[:, 0], X[:, 1], c=colors)
plt.show()
```

```python
# fit SVM
svm = LinearSVC(penalty='l2', loss='hinge', C=1, dual=True,
                random_state=42)
svm.fit(X, y)

(xx, yy) = np.meshgrid(np.arange(0, 5, 0.01), np.arange(0, 5, 0.01))
x = np.c_[xx.ravel(), yy.ravel()]
preds = svm.predict(x)
preds = preds.reshape(xx.shape)
plt.scatter(X[:, 0], X[:, 1], c=colors)
plt.contour(xx, yy, preds, [0.5])
```

```python
# fit SVM
svm = LinearSVC(penalty='l2', loss='hinge', C=100, dual=True,
                random_state=42)
svm.fit(X, y)

(xx, yy) = np.meshgrid(np.arange(0, 5, 0.01), np.arange(0, 5, 0.01))
x = np.c_[xx.ravel(), yy.ravel()]
preds = svm.predict(x)
preds = preds.reshape(xx.shape)
plt.scatter(X[:, 0], X[:, 1], c=colors)
plt.contour(xx, yy, preds, [0.5])
```

```python
def gaussian_kernel(sigma_squared):
    def kernel(X, Y):
        norm = np.zeros((len(X), len(Y)))
        for i in range(len(X)):
            for j in range(len(Y)):
                xi = X[i, :]
                xj = Y[j, :]
                diff = xi - xj
                norm[i, j] = diff.T.dot(diff)
        return np.exp(-norm / (2.0 * sigma_squared))
    return kernel

k = gaussian_kernel(4)
x1 = np.asarray([1, 2, 1]).reshape((1, -1))
x2 = np.asarray([0, 4, -1]).reshape((1, -1))
print(k(x1, x2))
```

```python
data = loadmat('ex6data2.mat')
X = data['X']
y = np.int32(data['y']).squeeze()
colors = ['white' if v == 1 else 'black' for v in y]
plt.scatter(X[:, 0], X[:, 1], c=colors)
plt.show()
```

```python
# fit SVM
k = gaussian_kernel(0.1 * 0.1)
svm = SVC(C=1, kernel=k, shrinking=False, random_state=42)
svm.fit(X, y)
print('ok')
(xx, yy) = np.meshgrid(np.arange(0, 1, 0.05), np.arange(0, 1, 0.05))
x = np.c_[xx.ravel(), yy.ravel()]
preds = svm.predict(x)
preds = preds.reshape(xx.shape)
plt.scatter(X[:, 0], X[:, 1], c=colors)
plt.contour(xx, yy, preds, [0.5])
```

```python
data = loadmat('ex6data3.mat')
X = data['X']
Xval = data['Xval']
y = np.int32(data['y']).squeeze()
yval = np.int32(data['yval']).squeeze()
colors = ['white' if v == 1 else 'black' for v in y]
plt.scatter(X[:, 0], X[:, 1], c=colors)
plt.show()
```

```python
values = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30]
results = []

for C in values:
    for sigma in values:
        k = gaussian_kernel(sigma * sigma)
        svm = SVC(C=C, kernel=k, shrinking=False, random_state=42)
        svm.fit(X, y)
        results.append((C, sigma, svm.score(Xval, yval)))
                       
results = sorted(results, key=lambda x: x[2], reverse=True)
for r in results[:10]:
    print(r)
```

```python
# fit SVM with the best parameters from the previous cell
k = gaussian_kernel(0.1 * 0.1)
svm = SVC(C=1, kernel=k, shrinking=False, random_state=42)
svm.fit(X, y)

(xx, yy) = np.meshgrid(np.arange(-1, 1, 0.05), np.arange(-1, 1, 0.05))
x = np.c_[xx.ravel(), yy.ravel()]
preds = svm.predict(x)
preds = preds.reshape(xx.shape)
plt.scatter(X[:, 0], X[:, 1], c=colors)
plt.contour(xx, yy, preds, [0.5])
```
