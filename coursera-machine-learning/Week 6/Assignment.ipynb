```python
% matplotlib inline
import matplotlib.pyplot as plt
from scipy.io import loadmat
import numpy as np
```

```python
mat = loadmat('ex5data1.mat')

X = mat['X']
Xtest = mat['Xtest']
Xval = mat['Xval']

y = mat['y']
ytest = mat['ytest']
yval = mat['yval']
```

Part 1

```python
plt.scatter(X, y)
plt.plot()
```

```python
def normalize(x, means=None, stds=None):
    x = x.copy()
    if means is None:
        means = x.mean(axis=0)
    if stds is None:
        stds = x.std(axis=0)
    for j in range(x.shape[1]):
        x[:, j] = (x[:, j] - means[j]) / stds[j]
    return (x, means, stds)
```

```python
def add_bias(x):
    bias = np.ones((x.shape[0], 1))
    return np.hstack([bias, x])
```

```python
train_x = X.copy()
#(train_x, means, stds) = normalize(X)
train_x = add_bias(train_x)
train_y = y.copy()
```

```python
def J(x, y, theta, l):
    m = float(x.shape[0])
    
    diff = x.dot(theta) - y
    cost = diff ** 2
    cost = (1.0 / (2.0 * m)) * cost.squeeze().sum()
    
    omega = (theta[1:] ** 2).sum()
    omega = (l / (2.0 * m)) * omega
    loss = cost + omega
    
    cost_grad = (1.0 / m) * x.T.dot(diff)
    omega_grad = np.zeros(theta.shape)
    omega_grad[1:] = (l / m) * theta[1:]
    gradients = cost_grad + omega_grad
    
    return (loss, gradients)
```

```python
theta = np.ones((2, 1))
l = 0
alpha = 0.001
n_iters = 5000

def optimize(x, y, alpha, theta, n_iters=5000, l=0):
    theta = theta.copy()
    for _ in range(n_iters):
        (loss, gradients) = J(x, y, theta, l)
        theta = theta - (alpha * gradients)
        #print(loss)
    return theta

theta = optimize(train_x, train_y, alpha, theta, n_iters, l)
```

```python
plot_x = [[np.min(train_x[:, 1])], [np.max(train_x[:, 1])]]
plot_x = np.asarray(plot_x)
plot_x = add_bias(plot_x)

plot_y = plot_x.dot(theta)

plot_x = plot_x[:, 1].squeeze().tolist()
plot_y = plot_y.squeeze().tolist()

plt.scatter(X, y)
plt.plot(plot_x, plot_y, color='r', linestyle='-')
```

Part 2

```python
def learning_curve(l, train_x, train_y, val_x, val_y):
    alpha = 0.001
    n_iters = 5000
    
    train_results = []
    val_results = []
    for n in range(1, len(train_x) + 1):
        theta = np.ones((train_x.shape[1], 1))
        x = train_x[:n]
        y = train_y[:n]
        theta = optimize(x, y, alpha, theta, n_iters, l)
        
        (train_error, _) = J(x, y, theta, 0)
        train_results.append((n, train_error))
        
        (val_error, _) = J(val_x, val_y, theta, 0)
        val_results.append((n, val_error))
        
    return (train_results, val_results)

val_x = Xval.copy()
val_x = add_bias(val_x)
val_y = yval.copy()

r = learning_curve(0, train_x, train_y, val_x, val_y)
(train_results, val_results) = r

plot_x = [d[0] for d in train_results]
plot_y = [d[1] for d in train_results]
plt.plot(plot_x, plot_y)

plot_x = [d[0] for d in val_results]
plot_y = [d[1] for d in val_results]
plt.plot(plot_x, plot_y)
```

```python
def polynomial_features(features, p):
    x = features.copy()
    for n in range(2, p + 1):
        x = np.hstack([x, features ** n])
    return x
```

Grid search for $\lambda$

```python
lambda_options = [0, 0.001, 0.003, 0.01, 0.03, 0.1, 0.3, 1, 3, 10]
def grid_search(train_x, train_y, val_x, val_y):
    alpha = 0.001
    n_iters = 5000
    
    train_results = []
    val_results = []
    for l in lambda_options:
        theta = np.ones((train_x.shape[1], 1))
        theta = optimize(train_x, train_y, alpha, theta, n_iters, l)
        
        (train_error, _) = J(train_x, train_y, theta, 0)
        train_results.append((l, train_error, theta))
        
        (val_error, _) = J(val_x, val_y, theta, 0)
        val_results.append((l, val_error, theta))
        
    return (train_results, val_results)
```

```python
r = grid_search(train_x, train_y, val_x, val_y)
(train_results, val_results) = r

plot_x = [d[0] for d in train_results]
plot_y = [d[1] for d in train_results]
plt.plot(plot_x, plot_y)

plot_x = [d[0] for d in val_results]
plot_y = [d[1] for d in val_results]
plt.plot(plot_x, plot_y)
```

```python
def random_learning_curve(l, train_x, train_y, val_x, val_y):
    alpha = 0.001
    n_iters = 5000
    n_samples = 20
    
    train_results = []
    val_results = []
    for n in range(1, len(train_x) + 1):
        train_error_sum = 0
        val_error_sum = 0
        for _ in range(n_samples):
            theta = np.ones((train_x.shape[1], 1))
            indices = np.arange(len(train_x))
            np.random.shuffle(indices)
            x = train_x[indices[:n]]
            y = train_y[indices[:n]]
            theta = optimize(x, y, alpha, theta, n_iters, l)
            (train_error, _) = J(x, y, theta, 0)
            train_error_sum += train_error
            
            indices = np.arange(len(val_x))
            np.random.shuffle(indices)
            x = val_x[indices[:n]]
            y = val_y[indices[:n]]
            (val_error, _) = J(x, y, theta, 0)
            val_error_sum += val_error
            
        train_results.append((n, train_error / n_samples))
        val_results.append((n, val_error / n_samples))
        
    return (train_results, val_results)
```

```python
l = 0.01
r = random_learning_curve(l, train_x, train_y, val_x, val_y)
(train_results, val_results) = r

plot_x = [d[0] for d in train_results]
plot_y = [d[1] for d in train_results]
plt.plot(plot_x, plot_y)

plot_x = [d[0] for d in val_results]
plot_y = [d[1] for d in val_results]
plt.plot(plot_x, plot_y)
```
