```python
X = raw_X.copy()
y = raw_y.copy()

bias = np.ones((X.shape[0], 1))
X = np.hstack([bias, X])

# number 0 has a label 10 in this dataset
y[y == 10] = 0
```

```python
def sigmoid(z):
    return 1.0 / (1 + np.exp(-z))

g = sigmoid
```

```python
def h(theta, X):
    return g(X.dot(theta))

def J(theta, X, y, l):
    (m, n) = X.shape[:2]
    h_theta_X = h(theta, X)
    y_1 = y.T.dot(np.log(h_theta_X))
    y_0 = (1.0 - y).T.dot(np.log(1.0 - h_theta_X))
    cost = (1.0 / m) * (-y_1[0, 0] - y_0[0, 0])

    reg = (l / (2.0 * m)) * np.square(theta[1:]).sum()
    
    return cost + reg 

def dJ_dthetaj(theta, X, y, l):
    (m, n) = X.shape[:2]
    cost = (1.0 / m) * X.T.dot(h(theta, X) - y)
    
    reg = np.zeros((n, 1))
    reg[1:] = theta[1:]
    reg = (l / m) * reg
    
    return cost + reg
```

```python
# test h, cost, and derivatives
theta = np.array([[-2], [-1], [1], [2]])

X_t = np.arange(1, 16).reshape((3, 5)) / 10.0
X_t = np.hstack([np.ones((5, 1)), X_t.T])

y_t = np.array([[1], [0], [1], [0], [1]])

l = 3.0

print('Cost', J(theta, X_t, y_t, l))
print('Gradients', dJ_dthetaj(theta, X_t, y_t, l))
```

```python
def learn_classifier(cls, theta):
    theta = theta.copy()
    ith_y = (y == cls)
    
    for _ in range(5000):
        theta = theta - alpha * dJ_dthetaj(theta, X, ith_y, l)
    
    loss = J(theta, X, ith_y, l)
    print('Final loss for {}th class = {:.3f}'.format(i, loss))
    
    return theta
```

```python
(m, n) = X.shape[:2]
n_classes = 10
theta = np.zeros((n_classes, X.shape[1]))
l = 0.1
alpha = 0.1

for i in range(n_classes):
    print('Learning classifier for {}th class'.format(i))
    ith_theta = theta[i, :].T.reshape(-1, 1)
    ith_theta = learn_classifier(i, ith_theta)
    theta[i, :] = ith_theta.T
```

```python
predictions = g(X.dot(theta.T))

y_hat = np.argmax(predictions, axis=1)
correct = float((y_hat == y[:, 0]).sum())
print('Accuracy {:.3f}'.format(correct / m))
```
