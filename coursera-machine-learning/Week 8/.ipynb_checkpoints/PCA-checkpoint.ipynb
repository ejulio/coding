```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

```python
data = loadmat('ex7data1.mat')
X = data['X']
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
# normalize data
means = X.mean(axis=0)
stds = X.std(axis=0)
X_norm = X.copy()

for j in range(X.shape[1]):
    X_norm[:, j] = (X[:, j] - means[j]) / stds[j]
```

```python
def pca(X):
    SIGMA = (1.0 / len(X)) * X.T.dot(X)
    (U, S, V) = np.linalg.svd(SIGMA)
    
    return (U, S)
```

```python
(U, S) = pca(X_norm)

y = means + 1.5 * S[0] * U[:, 0]
p1 = [means[0], y[0]]
p2 = [means[1], y[1]]
plt.plot(p1, p2)

y = means + 1.5 * S[1] * U[:, 1]
p1 = [means[0], y[0]]
p2 = [means[1], y[1]]
plt.plot(p1, p2)

plt.scatter(X[:, 0], X[:, 1])

plt.show()
```
