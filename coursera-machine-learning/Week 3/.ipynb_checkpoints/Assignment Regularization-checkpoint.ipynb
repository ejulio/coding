```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
raw_X = []
raw_y = []

with open('./ex2data2.txt', 'r') as file:
    content = file.read().splitlines()
    
content = map(lambda s: s.split(','), content)
for (score1, score2, result) in content:
    raw_X.append((float(score1), float(score2)))
    raw_y.append(int(result))
    
raw_X = np.asarray(raw_X)
raw_y = np.asarray(raw_y)
```

```python
plt.scatter(raw_X[:, 0], raw_X[:, 1], c=raw_y)
```

```python
def map_features(x, degree=6):
    out = []
    for d in range(1, degree + 1):
        for j in range(d + 1):
            f = (x[0] ** (d - j)) * (x[1] ** j)
            out.append(f)
    
    return np.asarray(out)
```

```python
m = len(raw_X)

X = []
for i in range(len(raw_X)):
    X.append(map_features(raw_X[i, :], 6))
    
X = np.hstack([np.ones((m, 1)), X])
y = raw_y.reshape((m, 1))
```

```python
def sigmoid(z):
    return 1.0 / (1 + np.exp(-z))

g = sigmoid
```

```python
def J(h_theta_x, y, l, theta):
    m = len(y)
    
    y1 = y.T.dot(np.log(h_theta_x))
    y0 = (1 - y).T.dot(np.log(1 - h_theta_x))
    cost = (-1.0 / m) * np.sum([y1, y0])
    
    reg = (l / (2.0 * m)) * np.sum(theta[1:] ** 2)
    
    return cost + reg
    
```

```python
def dJ_d(X, y, h_theta_x, l, theta):
    m = len(y)
    d_dcost = (1.0 / m) * X.T.dot(h_theta_x - y)
    d_dreg = (l / m) * theta
    d_dreg[0, 0] = 0
    return d_dcost + d_dreg
```

```python
theta = np.zeros((X.shape[1], 1))
h_theta_x = g(X.dot(theta))
l = 0.1
print(J(h_theta_x, y, l ,theta))
print(dJ_d(X, y, h_theta_x, l, theta))
```

```python
theta = np.zeros((X.shape[1], 1))
n_iters = 10000
alpha = 0.0001
l = 1

for i in range(n_iters):
    h_theta_X = g(X.dot(theta))
    loss = J(h_theta_X, y, l, theta)
    print('Loss: {:.5f}'.format(loss))
    theta = theta - (alpha * d_dtheta(X, y, h_theta_X, l, theta))
```

```python
score1 = np.linspace(-1, 1.5, 50)
score2 = np.linspace(-1, 1.5, 50)

result = np.zeros((50, 50))
for i in range(len(score1)):
    for j in range(len(score2)):
        x = map_features([score1[i], score2[j]])
        x = np.hstack([[1], x])
        x = x.reshape((len(x), 1)).T
#         normalize(x)
        result[i, j] = g(x.dot(theta))

plt.scatter(raw_X[:, 0], raw_X[:, 1], c=raw_y)
plt.contour(score1, score2, result, [0.5])
```
