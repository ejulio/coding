```python
# There are a couple more columns in the research file than in the general one
# lets check which ones
for column in df_research.columns:
    if column not in df_general.columns:
        print(column)
```

```python
# Tlet's check if there are columns that only exist in general payments, but not in research ones
for column in df_general.columns:
    if column not in df_research.columns:
        print(column)
```

```python
# the columns that only exist in one dataset be dropped
# otherwise they could be proxies to predicting if it was a payment to a research program or not
columns_to_drop = [c for c in df_research.columns if c not in df_general.columns]
df_research.drop(columns=columns_to_drop, inplace=True)

columns_to_drop = [c for c in df_general.columns if c not in df_research.columns]
df_general.drop(columns=columns_to_drop, inplace=True)
```

```python
print('### GENERAL ###')
print(df_general.Covered_Recipient_Type.value_counts())
print('### RESEARCH ###')
print(df_research.Covered_Recipient_Type.value_counts())

# both "Covered Recipient Teaching Hospital" and "Covered Recipient Physician" appear in General and Research
# however, "Covered Recipient Teaching Hospital" is more common in Research
```

```python
def unique_names(names):
    return set(n.lower() for n in names if isinstance(n, str))

unique_names(df_research.Teaching_Hospital_Name.values) & unique_names(df_general.Teaching_Hospital_Name.values)

# there is an overlap in "Teaching_Hospital_Name" for both General and Research
# so, this is not a proxy to Research Hospitals
```
