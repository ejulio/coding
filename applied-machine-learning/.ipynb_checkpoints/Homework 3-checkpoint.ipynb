```python
import pandas as pd
```

```python
df_general = pd.read_csv('./data/OP_DTL_GNRL_PGYR2017_P06282019.csv', nrows=50000)
```

```python
df_general.info()
```

```python
df_research = pd.read_csv('./data/OP_DTL_RSRCH_PGYR2017_P06282019.csv', nrows=50000)
```

```python
df_research.info(max_cols=200)
```

```python
# There are a couple more columns in the research file than in the general one
# lets check which ones
for column in df_research.columns:
    if column not in df_general.columns:
        print(column)
```

```python
# Tlet's check if there are columns that only exist in general payments, but not in research ones
for column in df_general.columns:
    if column not in df_research.columns:
        print(column)
```

```python
# the columns that only exist in one dataset be dropped
# otherwise they could be proxies to predicting if it was a payment to a research program or not
columns_to_drop = [c for c in df_research.columns if c not in df_general.columns]
df_research.drop(columns=columns_to_drop, inplace=True)

columns_to_drop = [c for c in df_general.columns if c not in df_research.columns]
df_general.drop(columns=columns_to_drop, inplace=True)
```

```python
df_research.info()
```

```python
df_general.info()
```

# Check the remaining columns

```python
print('### GENERAL ###')
print(df_general.Covered_Recipient_Type.value_counts())
print('### RESEARCH ###')
print(df_research.Covered_Recipient_Type.value_counts())

# both "Covered Recipient Teaching Hospital" and "Covered Recipient Physician" appear in General and Research
# however, "Covered Recipient Teaching Hospital" is more common in Research
```

```python
def unique_names(names):
    return set(n.lower() for n in names if isinstance(n, str))

unique_names(df_research.Teaching_Hospital_Name.values) & unique_names(df_general.Teaching_Hospital_Name.values)

# there is an overlap in "Teaching_Hospital_Name" for both General and Research
# so, this is not a proxy to Research Hospitals
```

```python
# Data related to Physician Names might not be important in a fist moment
# Physician_First_Name, Physician_Middle_Name, Physician_Last_Name, Physician_Name_Suffix

# Also, Data related to the address may not be important at first
# Recipient_Country, Recipient_Primary_Business_Street_Address_Line1, Recipient_Primary_Business_Street_Address_Line2
# Recipient_Zip_Code, Recipient_Postal_Code, Recipient_Province
```

```python
print(df_general.Recipient_State.value_counts())
print(df_research.Recipient_State.value_counts())

# Recipient_State and Recipient_City are related to each other and may be related to the Hospital
# They may be colinear
# Also, the distributions seem to change from General to Research
```

```python
print('### GENERAL ###')
print(df_general.Physician_Primary_Type.value_counts())
print('### RESEARCH ###')
print(df_research.Physician_Primary_Type.value_counts())

# This may be a proxy, leak
# If "Physician_Primary_Type" is "Chiropractor", then it is General
```

```python
print('### GENERAL ###')
print(df_general.Physician_Specialty.value_counts())
print('### RESEARCH ###')
print(df_research.Physician_Specialty.value_counts())

print('### INTERSECTION ###')
print(unique_names(df_research.Physician_Specialty.values) & unique_names(df_general.Physician_Specialty.values))

# The distribution of specialties seem to change from General to Research
```
