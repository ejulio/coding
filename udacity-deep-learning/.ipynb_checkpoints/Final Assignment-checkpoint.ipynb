Reading SVHN data...

```python
base_path = 'data/train/'
data_file_path = path.join(base_path, 'digits_data.pickle')
data = None

if path.exists(data_file_path):
    print('Loading data from Pickle')
    with open(data_file_path, 'rb') as file:
        data = pickle.load(file)
else:
    # some help from
    # https://stackoverflow.com/questions/33509163/how-to-read-mat-v7-3-files-in-python
    def get_attrs(name, obj):
        vals = []
        if obj.shape[0] == 1:
            vals.append(int(obj[0][0]))
        else:
            for k in range(obj.shape[0]):
                vals.append(int(data_file[obj[k][0]][0][0]))
        data[name].append(vals)
    
    print('Reading data from .mat file')
    matlab_data_file_path = path.join(base_path, 'digitStruct.mat')
    data_file = h5py.File(matlab_data_file_path)
    
    data = {}
    data['paths'] = []
    data['label'] = []
    data['height'] = []
    data['left'] = []
    data['top'] = []
    data['width'] = []
    
    print('Reading images names...')
    column = data_file['digitStruct']['name']
    for item in column:
        image_name = map(chr, data_file[item[0]][:])
        image_path = path.join(base_path, ''.join(image_name))
        data['paths'].append(image_path)
        
    print('Reading images bounding boxes and labels...')
    column = data_file['digitStruct']['bbox']
    for item in column:
        data_file[item[0]].visititems(get_attrs)
        
    print('Dumping Pickle file')
    with open(data_file_path, 'wb') as file:
        pickle.dump(data, file, pickle.HIGHEST_PROTOCOL)

print('DONE')
```

```python
def plot_bounding_box(data, index, ax):
    img = imread(data['paths'][index])
    left = np.min(data['left'][index])
    top = np.min(data['top'][index])
    bottoms = [sum(x) for x in zip(data['top'][index], data['height'][index])]
    bottom = np.max(bottoms)
    right = left + np.sum(data['width'][index])
    (rr, cc) = polygon_perimeter((top, bottom, bottom, top),
                                 (left, left, right, right),
                                 shape=img.shape)
    img[rr, cc] = 0
    ax.axis('off')
    ax.set_title(data['paths'][index])
    ax.imshow(img)


n_rows = 2
n_cols = 2
(fig, axes) = plt.subplots(nrows=n_rows, ncols=n_cols)

indexes = np.arange(len(data['paths']))
np.random.shuffle(indexes)

for i in range(n_rows):
    for j in range(n_cols):
        index = i * n_rows + j
        plot_bounding_box(data, indexes[index], axes[i, j])
```

```python
train_size = 27000
batch_size = 128
epochs = 10
iterations_per_epoch = int(train_size / float(batch_size))

height = 40
width = 90
channels = 1

n_classes = 10
n_chars_per_image = 6

data_indices = np.arange(len(data['paths']))
np.random.shuffle(data_indices)
train_indices = data_indices[:train_size]
validation_indices = data_indices[train_size:]
```

```python
class Transformer(object):
    _height = 0
    _width = 0
    
    def __init__(self, width, height):
        self._width = width
        self._height = height
        
    def transform(self, image):
        new_image = skimage.exposure.equalize_hist(image)
        shape = (self._height, self._width)
        new_image = skimage.transform.resize(image, shape)
        return (new_image / new_image.mean()) * new_image.std()
        
transformer = Transformer(width, height)
```

```python
class DataLoader(object):
    _data = None
    _transformer = None
    _batch_shape = 0
    _cursor = 0
    
    def __init__(self, batch_shape, data, data_indices, transformer):
        self._batch_shape = batch_shape
        self._data = data
        self._transformer = transformer
        self._indices = data_indices
        self._cursor = 0
        
    def shuffle(self):
        np.random.shuffle(self._indices)
        self._cursor = 0
    
    def next_batch(self):
        batch_size = self._batch_shape[0]
        images = np.zeros(self._batch_shape, dtype=np.float32)
        labels = np.zeros((batch_size, n_chars_per_image, n_classes), dtype=np.float32)
        for i in range(batch_size):
            self._cursor = (self._cursor + 1) % len(self._indices)
            index = self._indices[self._cursor]
            image = self._load_image(index)
            image = self._preprocess_image(image)
            images[i] = image[:, :, np.newaxis]
            
            label = self._get_label(index)
            label = self._preprocess_label(label)
            labels[i] = label
            
            
        return (images, labels)
    
    def _load_image(self, index):
        image = imread(self._data['paths'][index], as_grey=True)
        return skimage.img_as_float(image)
    
    def _preprocess_image(self, image):
        return self._transformer.transform(image)
    
    def _get_label(self, index):
        labels = self._data['label'][index]
        one_hot_labels = np.zeros((len(labels), n_classes))
        for (i, label) in enumerate(labels):
            # number 0 has label number 10 
            label = label % n_classes
            one_hot_labels[i, label] = 1
            
        return one_hot_labels
    
    def _preprocess_label(self, label):
        diff = n_chars_per_image - label.shape[0]
        padding = np.zeros((diff, n_classes), dtype=np.float32)
        padding[:, 0] = 1
        return np.vstack([padding, label])
```

```python
def conv2d_relu(data, feature_maps):
    conv_shape = (3, 3, data.shape[3].value, feature_maps)
    weights = tf.Variable(tf.truncated_normal(conv_shape, stddev=0.1))
    biases = tf.Variable(tf.zeros([feature_maps]))
    
    conv = tf.nn.conv2d(data, weights, strides=(1, 1, 1, 1), padding='SAME')
    return tf.nn.relu(conv + biases)


def max_pool(data):
    return tf.nn.max_pool(data,
                          ksize=(1, 2, 2, 1),
                          strides=(1, 2, 2, 1),
                          padding='SAME')

def fully_connected_relu(data, n_units):
    weights_shape = (data.shape[1].value, n_units)
    weights = tf.Variable(tf.truncated_normal(weights_shape, stddev=0.01))
    biases = tf.Variable(tf.constant(1.0, shape=[n_units]))
    
    hidden = tf.matmul(data, weights)
    return tf.nn.relu(hidden + biases)

graph = tf.Graph()
with graph.as_default():
    x = tf.placeholder(tf.float32,
                       shape=(None, height, width, channels))
    y = tf.placeholder(tf.float32,
                       shape=(None, n_chars_per_image, n_classes))
    keep_prob = tf.placeholder(tf.float32)
    
    conv1 = conv2d_relu(x, 16)
    conv2 = conv2d_relu(conv1, 16)
    conv2 = max_pool(conv2)
    
    conv3 = conv2d_relu(conv2, 32)
    conv4 = conv2d_relu(conv3, 32)
    conv4 = max_pool(conv4)
    
    conv5 = conv2d_relu(conv4, 64)
    conv5 = max_pool(conv5)
    
    n_units = conv5.shape[1].value * conv5.shape[2].value * conv5.shape[3].value
    reshape = tf.reshape(conv5, [-1, n_units])
    
    fc1 = fully_connected_relu(reshape, 512)
    fc1 = tf.nn.dropout(fc1, keep_prob)
    fc2 = fully_connected_relu(fc1, 256)
    
    losses_sum = None
    softmaxes = []
    for char_index in range(n_chars_per_image):
        ith_logits = fully_connected_relu(fc2, n_classes)
        softmaxes.append(tf.nn.softmax(ith_logits))
        ith_loss = tf.nn.softmax_cross_entropy_with_logits(labels=y[:, char_index, :],
                                                           logits=ith_logits)
        if losses_sum is None:
            losses_sum = ith_loss
        else:
            losses_sum = losses_sum + ith_loss
        
    loss = tf.reduce_mean(losses_sum)
    
    optimizer_iteration = tf.Variable(0)
    learning_rate = tf.train.exponential_decay(0.1,
                                               optimizer_iteration, 
                                               decay_steps=520,
                                               decay_rate = 0.96)
    optimizer = tf.train.GradientDescentOptimizer(learning_rate)
    optimizer = optimizer.minimize(loss, global_step=optimizer_iteration)
```

```python
train_loader = DataLoader(batch_shape=(batch_size, height, width, channels),
                          data=data,
                          data_indices=train_indices,
                          transformer=transformer)

valid_loader = DataLoader(batch_shape=(batch_size, height, width, channels),
                          data=data,
                          data_indices=validation_indices,
                          transformer=transformer)

with tf.Session(graph=graph) as session:
    session.run(tf.global_variables_initializer())
    for epoch in range(epochs):
        train_loader.shuffle()
        
        for iteration in range(iterations_per_epoch):
            (images, labels) = train_loader.next_batch()
            feed_dict = {x: images,
                         y: labels,
                         keep_prob: 0.5}
            (_, l) = session.run([optimizer, loss], feed_dict=feed_dict)
            
            print('Loss at {}/{} of {} = {:.05f}.'.format(epoch, iteration, iterations_per_epoch, l))
```
